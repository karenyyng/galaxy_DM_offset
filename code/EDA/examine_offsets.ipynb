```python
offset_list = np.array(pickle.load(open('offset_list_129.pkl')))
relaxedness_list = np.array(pickle.load(open('relaxedness_129.pkl')))
```

```python
_, _, _ = plt.hist(offset_list[:, 0], normed=True, range=(0, 2e3),
                   bins=30)
plt.title('KDE peak offset in c kpc/h')
plt.xlabel('offset in c kpc/h')
```

```python
_, _, _ = plt.hist(offset_list[:, 1],
                   bins=50)
plt.title('KDE peak offset in fraction of R200C')
plt.xlabel('offset in terms of % of R200C')
plt.ylabel('Number of clusters')
plt.savefig('../plots/KDE_peak_mostBound_offset_full_range.png', bbox_inches='tight')
```

```python
_, _, _ = plt.hist(offset_list[:, 1], normed=True,
                   bins=50, range=(0, 10))
plt.title('KDE peak offset in fraction of R200C')
plt.xlabel('offset in terms of % of R200C')
plt.ylabel('PDF')
plt.savefig('../plots/KDE_peak_mostBound_offset_zoomed.png', bbox_inches='tight')
```

```python
_, _, _ = plt.hist(relaxedness_list[mask], histtype='step',
                   label='offset > 10%')
_, _, _ = plt.hist(relaxedness_list[~mask], histtype='step', color='r',
                   label='offset < 10%')
plt.title('clusters')
plt.xlabel('galaxy-most bound particle offset in terms % of R200C')
plt.ylabel('number of clusters')
plt.legend(loc='best')
plt.savefig('../plots/why_huge_offset.png',
            bbox_inches='tight')
```
