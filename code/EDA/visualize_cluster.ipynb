# compute $d(CM_{no wrap}, mostBoundParticle)$

the CM and mostBoundParticle has the same dimensions we 
can just subtract element by element wise 

all the most bound particles are positive 

```python
rawdistNoWrap = CM - mostBoundParticle
rawdistNoWrap.shape  
```

examine raw dist in each spatial dim. for the first cluster

```python
rawdistNoWrap[:, 0]  # in kpc / h 
```

the numbers are huge in each spatial dim.
Now let's compute the magitude of the dist for each cluster.   
Take square root of the dot product between the difference of CM and mostBound positions
for each cluster 

```python
distCMmostBound = \
    np.array([np.sqrt(np.dot(rawdistNoWrap[:, i], rawdistNoWrap[:, i]))
              for i in range(rawdistNoWrap.shape[1])])
```

divide by R200C to see the ratio

```python
_, _, _ = plt.hist(distCMmostBound / \
                   f["Group"]["Group_R_Crit200"][:massGr1e13], bins=30)
plt.ylabel('freq')
plt.xlabel('d(CM, mostbound) / R200C')
```
