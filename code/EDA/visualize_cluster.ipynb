# plot 

```python
h5File = \
    "../data/" + \
    "Illustris-1_fof_subhalo_myCompleteHaloCatalog_00135.hdf5"
f = h5py.File(h5File, "r")
```

```python
df_list = [ec.extract_clst(f, clstNo) for clstNo in range(12)]
relaxedness1_list = \
    [cp.compute_relaxedness1(df_list[clstNo], f, clstNo) 
     for clstNo in range(12)]
```

```python
for i in range(len(df_list)):
    df = df_list[i]
    plt.axes().set_aspect('equal')
    plt.title('cluster {0}'.format(i + 1) + 
              ', relaxedness: {0:.0f}'.format(relaxedness1_list[i]))
    plt.plot(df["SubhaloPos0"], 
            df["SubhaloPos1"], 'r.', alpha=0.1)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    plt.close()
    plt.axes().set_aspect('equal')
    plt.title('cluster {0}'.format(i + 1))
    plt.plot(df["SubhaloPos0"], 
             df["SubhaloPos2"], '.', alpha=0.1)
    plt.show()
    plt.close()
    plt.axes().set_aspect('equal')
    plt.title('cluster {0}'.format(i + 1))
    plt.plot(df["SubhaloPos1"], 
             df["SubhaloPos2"], '.', alpha=0.1)
    plt.show()
    plt.close()
    
```
